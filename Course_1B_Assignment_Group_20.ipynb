# First Case Study : Miles Per Gallon

- Drivers are concerned with the mileage obtained by their automobiles. One driver has kept track of several tankfuls of gasoline
by recording miles driven and gallons used for each tankful. Develop a sentinel-controlled-repetition script that prompts the
user to input the miles driven and gallons used for each tankful.

- The script should calculate and display the miles per gallon obtained for each tankful. After processing all input information, the
script should calculate and display the combined miles per gallon obtained for all tankfuls (that is, total miles driven divided by
total gallons used).

- Requirement Output of the program: image.png

```python
def main():
    total_miles = 0.0
    total_gallons = 0.0
    tank_count = 0
    
    while True:
        try:
            gallons = float(input("Enter the gallons used (enter -1 to end): "))
            if gallons == -1:
                break
            miles = float(input("Enter the miles driven: "))
            
            if gallons <= 0:
                print("Error: Gallons used must be greater than zero. Please try again.")
                continue
            
            mpg = miles / gallons
            print(f"Miles per gallon for this tankful: {mpg:.2f}")
            
            total_miles += miles
            total_gallons += gallons
            tank_count += 1
        
        except ValueError:
            print("Error: Invalid input. Please enter a valid number.")

    overall_mpg = total_miles / total_gallons if total_gallons != 0 else 0.0
    print(f"\nOverall miles per gallon for {tank_count} tankful(s): {overall_mpg:.2f}")

if __name__ == "__main__":
    main()
```

# Second Case Study : A Game of Chance - Crap

- You roll two six-sided dice, each with faces containing one, two, three, four, five and six spots, respectively. \
When the dice come to rest, the sum of the spots on the two upward faces is calculated.

- If the sum is 7 or 11 on the first roll, you win. If the sum is 2, 3 or 12 on the first roll (called “craps”), you lose
(i.e., the “house” wins).

- If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your “point.” To win, you must continue rolling the dice until you “make your point” (i.e., roll that same point value). You lose by rolling a 7 before making your point.



```python
import random
def roll_dice(n_sides):
    return random.randint(1,n_sides)

def craps():
  # First rolls
  dice_1 = roll_dice(6)
  dice_2 = roll_dice(6)
  print(f"The first rolls are {dice_1} and {dice_2}.")


  sum = dice_1 + dice_2
  print("Sum is", sum)
  if sum == 7 or sum == 11: # Instant Win Conditions
    print("You win!")
    return
  elif sum == 2 or sum == 3 or sum == 12: # Instant Lose Conditions
    print("You lose!")
    return

  point = sum
  print(f"Point is {point}")
  sum = 0

  input("Press Enter to roll dices")
  while sum != point: # Roll until sum is equal to point to win (Get the same sum again from the first roll to win)
    dice_1 = roll_dice(6)
    dice_2 = roll_dice(6)
    sum = dice_1 + dice_2
    print(f"The next roll is {dice_1} and {dice_2}.")
    print("Sum is", sum)
    if sum == 7:  # But if sum is 7, lose.
      print("You lose!")
      return
    input("Press Enter to roll dices")
  print("You win!")

craps()


```

# Third Case Study : COVID 19 Infection Statistic

- (COVID-19 Infection Statistics) During the first 20 days of the COVID-19 pandemic, the number of newly infected patients per day in a country were recorded.

- Place the following numbers in a list: 174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301, 1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342 Use the built-in functions in the statistics module to display the following statistics concerning the infection rate: minimum, maximum, range, mean, median, variance, and standard deviation.



```python
import statistics

n_infected_patients = [174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301, 1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342]

min_value = min(n_infected_patients)
max_value = max(n_infected_patients)
range_value = max_value - min_value
mean_value = sum(n_infected_patients) / len(n_infected_patients)
median_value = statistics.median(n_infected_patients)
variance_value = statistics.variance(n_infected_patients)
std_dev_value = statistics.stdev(n_infected_patients)

print("The minimum number of patients infected in a day was", min_value)
print("The maximum number of patients infected in a day was", max_value)
print("The range of patients infected in a day was", range_value)
print("The mean number of patients infected in a day was", mean_value)
print("The median number of patients infected in a day was", median_value)
print("The variance in the number of patients infected in a day was", variance_value)
print("The standard deviation in the number of patients infected in a day was", std_dev_value)
```
