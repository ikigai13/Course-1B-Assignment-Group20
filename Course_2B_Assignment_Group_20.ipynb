### Checking and Handling Null Values

```python
# Check the data set if theres any null/missing value
print(df.isnull().sum())
```

```python
null_records = df[df['product_weight_gram'].isnull()]

print(null_records.head())
```

### Correlation?

```python
# Calculate the correlation matrix
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include='number')

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### Comparation Between Price and Order Status

```python
# Check the distribution of order statuses
order_status_counts = df['order_status'].value_counts()

# Print counts of each order status
print("Order Status Counts:")
print(order_status_counts)

# Visualize the distribution of order statuses
plt.figure(figsize=(8, 5))
sns.countplot(x='order_status', data=df, palette='viridis')
plt.title('Distribution of Order Statuses')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='order_status', y='price', data=df)
plt.title('Price Distribution by Order Status')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import f_oneway

# ANOVA: Compare prices across different order statuses
order_status_groups = [df[df['order_status'] == status]['price'] for status in df['order_status'].unique()]
f_stat, p_value = f_oneway(*order_status_groups)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

F-stat â‰ˆ 1, variance between groups not significantly larger than variance within each group

P-value >= 0.05, insufficient evidence to conclude that there is a significant difference in prices across different order statuses.

```python
def plot_kde(column_name):
    plt.figure(figsize=(12, 6))
    sns.kdeplot(df[column_name], shade=True)
    plt.title(f'Distribution for {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Density')
    plt.show()
```

### Top 10 Products

```python
product_quantity_sum = df.groupby('product_id')['quantity'].sum()

# Sort products based on the sum of quantity and select top 10
top_products = product_quantity_sum.sort_values(ascending=False).head(10)

# Merge with the original DataFrame to get category information
top_products_info = df[df['product_id'].isin(top_products.index)]
top_products_info = top_products_info.drop_duplicates(subset=['product_id'])  # Drop duplicates to get unique product info

# Plot the top 10 products with annotations
plt.figure(figsize=(10, 6))
bars = plt.bar(top_products.index.astype(str), top_products.values, color='skyblue')

# Annotate each bar with its product category
for bar, product_id in zip(bars, top_products.index):
    category = top_products_info[top_products_info['product_id'] == product_id]['product_category_name'].iloc[0]
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, category,
             ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Product ID')
plt.ylabel('Total Quantity')
plt.title('Top 10 Products by Total Quantity with Categories')
plt.xticks(rotation=45)

# Adjust x-axis ticks and labels to start from 0
plt.gca().set_xticks(range(len(top_products.index)))
plt.gca().set_xticklabels(top_products.index.astype(str))
plt.tight_layout()
plt.show()
```

```python
product_price_sum = df.groupby('product_id')['price'].sum()

# Sort products based on the total price and select top 10
top_products_price = product_price_sum.sort_values(ascending=False).head(10)

# Merge with the original DataFrame to get category information
top_products_info = df[df['product_id'].isin(top_products_price.index)]
top_products_info = top_products_info.drop_duplicates(subset=['product_id'])  # Drop duplicates to get unique product info

# Plot the top 10 products by total price with annotations
plt.figure(figsize=(10, 6))
bars = plt.bar(top_products_price.index.astype(str), top_products_price.values, color='skyblue')

# Annotate each bar with its product category
for bar, product_id in zip(bars, top_products_price.index):
    category = top_products_info[top_products_info['product_id'] == product_id]['product_category_name'].iloc[0]
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100000, category,
             ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Product ID')
plt.ylabel('Total Price (Rp)')
plt.title('Top 10 Products by Total Price with Categories')
plt.xticks(rotation=45)

# Adjust x-axis ticks and labels to start from 0
plt.gca().set_xticks(range(len(top_products_price.index)))
plt.gca().set_xticklabels(top_products_price.index.astype(str))
plt.tight_layout()
plt.show()
```

```python
df['normalized_price'] = df['price'] / df['quantity']
product_normalized_price_sum = df.groupby('product_id')['normalized_price'].sum()

# Sort products based on the normalized price and select top 10
top_products_normalized_price = product_normalized_price_sum.sort_values(ascending=False).head(10)

# Merge with the original DataFrame to get category information
top_products_info_normalized = df[df['product_id'].isin(top_products_normalized_price.index)]
top_products_info_normalized = top_products_info_normalized.drop_duplicates(subset=['product_id'])  # Drop duplicates to get unique product info

# Plot the top 10 products by normalized price with annotations
plt.figure(figsize=(10, 6))
bars = plt.bar(top_products_normalized_price.index.astype(str), top_products_normalized_price.values, color='lightgreen')

# Annotate each bar with its product category
for bar, product_id in zip(bars, top_products_normalized_price.index):
    category = top_products_info_normalized[top_products_info_normalized['product_id'] == product_id]['product_category_name'].iloc[0]
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, category,
             ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Product ID')
plt.ylabel('Normalized Price (Rp)')
plt.title('Top 10 Products by Normalized Price with Categories')
plt.xticks(rotation=45)

# Adjust x-axis ticks and labels to start from 0
plt.gca().set_xticks(range(len(top_products_normalized_price.index)))
plt.gca().set_xticklabels(top_products_normalized_price.index.astype(str))
plt.tight_layout()
plt.show()
```

```python
category_quantity_sum = df.groupby('product_category_name')['quantity'].sum()

# Sort categories based on the sum of quantity and select top 10
top_categories = category_quantity_sum.sort_values(ascending=False).head(10)

# Plot the top 10 categories
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity')
plt.title('Top 10 Product Categories by Total Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Group By Customer ID

```python
# Filter the DataFrame to include only canceled orders
canceled_orders = df[df['order_status'] == 'canceled']

# Group by 'customer_id' and count the number of canceled orders
top_canceling_customers = canceled_orders.groupby('customer_id').size().sort_values(ascending=False).head(10)

# Display the top 10 customers who cancel the most orders
print("Top 10 Customers with the Most Canceled Orders:")
print(top_canceling_customers)

# Plotting (optional)
plt.figure(figsize=(10, 6))
top_canceling_customers.plot(kind='bar', color='salmon')
plt.xlabel('Customer ID')
plt.ylabel('Number of Canceled Orders')
plt.title('Top 10 Customers with the Most Canceled Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Group By Seller ID

```python
# Filter the DataFrame to include only canceled orders
canceled_orders = df[df['order_status'] == 'canceled']

# Group by 'seller_id' and count the number of canceled orders
top_canceling_sellers = canceled_orders.groupby('seller_id').size().sort_values(ascending=False).head(10)

# Display the top 10 sellers who cancel the most orders
print("Top 10 Sellers with the Most Canceled Orders:")
print(top_canceling_sellers)

# Plotting (optional)
plt.figure(figsize=(10, 6))
top_canceling_sellers.plot(kind='bar', color='salmon')
plt.xlabel('Seller ID')
plt.ylabel('Number of Canceled Orders')
plt.title('Top 10 Seler with the Most Canceled Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Grouping by Payment Type

```python
group_by_payment = df.groupby('payment_type').agg({'order_id': 'count', 'price': 'sum'}).reset_index()
group_by_payment.rename(columns={'order_id': 'record_count'}, inplace=True)
```

```python
group_by_payment
```

```python
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import FuncFormatter
# Plot the count of records
plt.figure(figsize=(10, 5))
plt.bar(group_by_payment['payment_type'], group_by_payment['record_count'], color='tab:blue', alpha=0.6)
plt.xlabel('Payment Type')
plt.ylabel('Record Count')
plt.title('Record Count by Payment Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
# Function to format the y-axis labels in millions
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fB' % (x * 1e-9)

# Plot the sum of prices in millions of Rupiah
plt.figure(figsize=(10, 5))
plt.bar(group_by_payment['payment_type'], group_by_payment['price'], color='tab:green', alpha=0.6)
plt.xlabel('Payment Type')
plt.ylabel('Total Price (Milyar Rp)')
plt.title('Total Price by Payment Type')

# Apply the formatter to the y-axis
formatter = FuncFormatter(millions)
plt.gca().yaxis.set_major_formatter(formatter)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

It seems that there are no significant differences in record count and sum of price between different payment types. (Is the dataset synthetic?)
